So what can we learn from this simple experiment?
==================================================

When we only take a few samples from the distribution then we see a lot of variability. As we take more and more samples, the parameters of the distribution we have measured look more and more like the distribution from which we are sampling.

So - when we take a sample we can only really recover the "true" underlying distribution when we take an infinite number of samples. With any finite number of samples we will not recover exactly the underlying true distribution.


**what does this mean in practice?**

Lets start with a simple question. What is the average number we would expect from a roll of the dice?

The answer is straightforward to calculate. For an unbiased dice every number is equally likely, therefore we would expect to to a 1 as often as a 6, a 2 as often as a 4.

Therefore over 6N rolls we would expect:
    
* $N$ 1s
* $N$ 2s
* $N$ 3s
* $N$ 4s
* $N$ 5s
* $N$ 6s

So the total pip count for $6N$ rolls would be:

$N + 2BN + 3N + .. + 6N = 21N$

Therefore the average score will be $\frac{21N}{6N} = 3.5$

So if we were to sample (roll the dice) n times we would expect an average score over the n rolls of 3.5.

So, lets set up a new random variable $ y_n $. We will define $y_n$ as the average score we get from rolling a dice $n$ times.

For example, we could set n = 10

A little script generate an array of 10 values and then get the average




```python
average_dicearray = average(random.random_integers(1,6,10000))
print average_dicearray
```

lets repeat this 100 time - in other words get 100 values of $y_{10}$

```python
rolls = 10
ntries = 100
average_dice = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice[i] = av

print average_dice
```

lets plot these values in a histogram:

```python
plt.hist(average_dice)
plt.title("Histogram showing average scores (10 rolls, 100 tries)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

What we see is a function centred on 3.5 - but with some spread around that.

What would happen if we looked at 10000 rolls what distribution would we see?

```python
rolls = 10
ntries = 10000
average_dice = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice[i] = av

plt.hist(average_dice)
plt.title("Histogram showing average scores (10 rolls, 10000 tries)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

what would happen if we looked at the average of 20 rolls at a time instead?

```python
rolls = 20
ntries = 10000
average_dice = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice[i] = av

plt.hist(average_dice)
plt.title("Histogram showing average scores (20 rolls, 10000 tries)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
It becomes a bit clear if we put them on the same graph:
```

```python
rolls = 10
ntries = 10000
average_dice1 = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice1[i] = av

rolls = 20
ntries = 10000
average_dice2 = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice2[i] = av

plt.hist(average_dice1, histtype = 'step', color = 'red')
plt.hist(average_dice2, histtype = 'step', color = 'blue')
plt.title("Histogram showing average scores (10 & 20 rolls, 10000 tries)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

It seems like the distribution where we took the distribution of the averages of 20 rolls (blue) is tighter than the distribution for 10 rolls(red). So lets extend the experiment and look at adding a new line (black) with 100 rolls each try. 

```python
rolls = 10
ntries = 10000
average_dice1 = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice1[i] = av

rolls = 20
ntries = 10000
average_dice2 = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice2[i] = av

rolls = 100
ntries = 10000
average_dice3 = zeros(ntries)
for i in xrange(ntries):
    av = average(random.random_integers(1,6,rolls))
    average_dice3[i] = av

   
plt.hist(average_dice1, histtype = 'step', color = 'red')
plt.hist(average_dice2, histtype = 'step', color = 'blue')
plt.hist(average_dice3, histtype = 'step', color = 'black')
plt.title("Histogram showing average scores (10, 20 and 100 rolls, 10000 tries)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

It would therefore seem that the more values we take each time, the closer we get to having an average that is predictably close to 3.5.

Lets look at this another way. We known the correct answer for the average is 3.5. Out of the 10000 tries, how many times are do we get an average of the rolls that is more than 0.2 away from the true value. i,e how many times is $y_n \lt 3.3$  or $y_n \gt 3.7$ for n = 10, 20 or 100?

```python
n1 = where(average_dice1 > 3.7,1,0)
n2 = where(average_dice2 > 3.7,1,0)
n3 = where(average_dice3 > 3.7,1,0)

n4 = where(average_dice1 < 3.3,1,0)
n5 = where(average_dice2 < 3.3,1,0)
n6 = where(average_dice3 < 3.3,1,0)

out10 = sum(n1)+sum(n4)
out20 = sum(n2)+sum(n5)
out100 = sum(n3)+sum(n6)

print out10, out20, out100
```

So you can see the more values we take, the less likely the value of $y_n$ is to differ too significantly from the true average
